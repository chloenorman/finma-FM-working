```python
# Exercise 8: Portfolio Performance Report
# Write your code here
import pandas as pd 
import numpy as np

transactions = pd.read_csv("transactions.csv")
stock_prices = pd.read_csv("stock_prices_timeseries.csv", index_col = "Date")
company_info = pd.read_csv("company_info.csv")

#current value
current_value=stock_prices.iloc[-1]
#print(current_value.T)

#cost basis
buy_comms = transactions.loc[transactions["action"]=="BUY"].groupby("symbol")["commission"].sum()
sell_comms = transactions.loc[transactions["action"]=="SELL"].groupby("symbol")["commission"].sum()
purchase_cost = transactions.loc[transactions["action"]=="BUY"].groupby("symbol")["shares"].sum()*transactions.loc[transactions["action"]=="BUY"].groupby("symbol")["price"].sum()
sale_cost = transactions.loc[transactions["action"]=="SELL"].groupby("symbol")["shares"].sum()*transactions.loc[transactions["action"]=="SELL"].groupby("symbol")["price"].sum()
p_cost_comm = purchase_cost + buy_comms
s_cost_comms = sale_cost - sell_comms
net_total_cost = p_cost_comm.subtract(s_cost_comms, fill_value=0)
print("the cost basis per holding is", (buy_comms+purchase_cost))

#unrealised profit/loss
B = transactions.loc[transactions["action"]=="BUY"].groupby("symbol")["shares"].sum()
S = transactions.loc[transactions["action"]=="SELL"].groupby("symbol")["shares"].sum()
net_holds = B.subtract(S, fill_value=0)
unrealised_pl = (current_value - (net_total_cost)/net_holds)*net_holds
print("the unrealised profit or loss for each holding is \n", unrealised_pl)
```
